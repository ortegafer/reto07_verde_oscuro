```python
d2.info()
fecha_referencia = pd.to_datetime('2025-02-11')
d2['Antigüedad'] = (fecha_referencia - d2['FECHA ENTRADA']).dt.days // 365
d2.head()
d2['Franja_Antigüedad'] = pd.cut(
    d2['Antigüedad'], 
    bins=[-1, 5, 10, 15, 20, 25], 
    labels=['0-5', '6-10', '11-15', '16-20', '21-25'], 
    right=True)
d2.head()
```

```python
# Definir categorías e indicadores ficticios
categories = ['A', 'B', 'C', 'D', 'E']
countries = ['País 1', 'País 2', 'País 3']  # Definir tres países
values_dict = {country: np.random.rand(len(categories)) for country in countries}  # Generar valores aleatorios para cada país

# Crear el radar plot
fig = go.Figure()

# Añadir un trazo para cada país
for country, values in values_dict.items():
    values = np.append(values, values[0])  # Cerrar el radar plot agregando el primer valor al final
    categories_with_first = categories + [categories[0]]  # Asegurarse de que la primera categoría se repita al final
    
    fig.add_trace(go.Scatterpolar(
        r=values,
        theta=categories_with_first,
        fill='toself',
        name=country
    ))

# Configurar el diseño del gráfico
fig.update_layout(polar=dict(radialaxis=dict(visible=True,range=[0,1])),title='Radar Plot por país'
)

# Mostrar el gráfico
fig.show()
```
